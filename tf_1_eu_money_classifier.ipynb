# **European Money Denominations Classifier**

Import dataset

*   !wget: download dataset  
  *   --quiet: surpress progress message
  *   -o: set output dir

*   !tar: command regarding archives operation
  *   -x: extract files
  *   -z: ensure archive to be decompressed is in gzip (*.gz*)
  *   -f: use specified archive file to be extracted
  *   -C: specify extract directory
  *   --exclude '.*': exclude files that starts with dot (.) to be extracted




Initiate directory for dataset

```python
import os

data_path = './data'

if not os.path.exists(data_path):
  os.makedirs(data_path)
```

Import training data

```python
fn_train = 'train_data_keras'
archive_ext = '.tar.gz'

download_path = os.path.join(data_path, fn_train)
train_path = os.path.join(data_path, 'train')

if not os.path.exists(train_path):
  os.makedirs(download_path)

  !wget --quiet -O {download_path + archive_ext} https://cocl.us/DL0320EN_TRAIN_TAR_KERAS
  !tar -xzf {download_path + archive_ext} -C {data_path} --exclude '.*'

  os.rename(download_path, train_path)
  os.remove(download_path + archive_ext)
```

Import validation data

```python
fn_validation = 'validation_data_keras'

download_path = os.path.join(data_path, fn_validation)
validation_path = os.path.join(data_path, 'validation')

if not os.path.exists(validation_path):
  os.makedirs(download_path)

  !wget --quiet -O {download_path + archive_ext} https://cocl.us/DL0320EN_VALID_TAR_KERAS
  !tar -xzf {download_path + archive_ext} -C {data_path} --exclude '.*'

  os.rename(download_path, validation_path)
  os.remove(download_path + archive_ext)
```

Import test data

```python
fn_test = 'test_data_keras'

download_path = os.path.join(data_path, fn_test)
test_path = os.path.join(data_path, 'test')

if not os.path.exists(test_path):
  os.makedirs(download_path)

  !wget --quiet -O {download_path + archive_ext} https://cocl.us/DL0320EN_TEST_TAR_KERAS
  !tar -xzf {download_path + archive_ext} -C {data_path} --exclude '.*'

  os.rename(download_path, test_path)
  os.remove(download_path + archive_ext)
```

Data exploration
